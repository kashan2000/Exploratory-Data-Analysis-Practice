#### One Hot Encoding for various features

```python
df= pd.read_csv("Mercedes data.csv",usecols=['X0','X1','X2','X4','X5'])
```

```python
df
```

```python
top_10=df['X0'].value_counts().sort_values(ascending=False).head(10).index
```

```python
top_10
```

```python
top_10= list(top_10)
```

```python
top_10
```

```python
for i in top_10:
    df[i]= np.where(df['X0']==i,1,0)
```

```python
#### done only for 1 coloumn
df.head(10)
```

#### Target Encoding

```python
df['Cabin']= df['Cabin'].astype(str).str[0]
```

```python
df['Cabin'].value_counts()
```

```python
df.head(20)
```

```python
labels= df.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
labels
```

```python
labels2= {i:k for k,i in enumerate(labels,0)}
```

```python
labels2
```

```python
df['Cabin']= df['Cabin'].map(labels2)
```

```python
df.head(10)
```

### The cabin having the highest survival rate is given the highest value in encoding
## This is How Target Ordinal Encoding is done
