MERCEDES DATA SET 
ONE HOT ENCODING TECHNIQUE

```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("Mercedes data.csv",usecols=['X1','X2','X3','X4','X5','X6'])
```

```python
df.head(5)
```

```python
for col in df.columns:
    print(col,':',len(df[col].unique()),'labels')
```

```python
#As we can see there are 27 unique labels in X1 and similarly in other categories.
```

```python
pd.get_dummies(df,drop_first=True).shape
```

```python
#As we can see a large number of columns are formed which can lead to curse of Dimentionallity
#So we will perform another meathod
```

```python
total= df['X2'].value_counts().sort_values(ascending=False).head(10)
```

```python
top_10 = [x for x in total.index]
```

```python
top_10
#Now we will perform One Hot Encoding to the first ten labels only
```

```python
for label in top_10:
    df[label]= np.where(df['X2']==label,1,0)
df[['X2']+top_10].head(40)
```

```python
def getting_dummy(data,variable,top_xlabels) :
    for label in top_xlabels:
        data[variable+'_'+label]= np.where(df[variable]==label,1,0)
```

```python
getting_dummy(df,'X2',top_10)
```
